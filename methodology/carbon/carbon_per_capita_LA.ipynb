```python
import pandas as pd
```

```python
# load dataset
DATA_PATH = "carbon_la_CO2_emission_estimates_2005-2019.csv"
df = pd.read_csv(DATA_PATH)
```

```python
# group per LA and sort by year
df_grouped = df.groupby('Local Authority').apply(lambda x: x.sort_values('Year'))
```

```python
# Calculate percentage difference within each group
df_grouped['Percentage Difference'] = df_grouped.groupby('Local Authority')['Per Capita Emissions (tonnes)'].pct_change() * 100
```

```python
# Drop NaN values resulting from the percentage difference calculation
df_grouped.dropna(subset=['Percentage Difference'], inplace=True)

# Reset the index
df_grouped.reset_index(drop=True, inplace=True)
```

```python
pivoted_df = pd.pivot_table(df_grouped, index='Local Authority', columns='Year', values=['Percentage Difference'])
pivoted_df
```

```python
# save to a file
pivoted_df.round(2).to_csv('carbon_per_capita_LA.csv')
```
